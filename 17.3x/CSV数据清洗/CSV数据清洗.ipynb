目标 1：实现 clean_csv_file 函数

参数：

input_path：一个字符串，表示原始 CSV 文件的绝对路径；
output_path：一个字符串，表示清洗后的 CSV 文件保存路径。
功能：

读取原始 CSV 文件；

删除完全重复的行；

对以下字段按指定策略填充缺失值：

字段名	填充策略
年龄	使用该列的均值填充
客户类型	使用该列的众数填充
注册时间	使用前向填充，即用同一列中前一条有效记录的值填充
消费金额	使用该列的中位数填充
所在城市	使用字符串 "未知" 填充

```python
import pandas as pd

def clean_csv_file(
    input_path: str,
    output_path: str,
) -> None:
    df = pd.read_csv(input_path)
    df.drop_duplicates(inplace=True)
    df['年龄']=df['年龄'].fillna(df['年龄'].mean())

    df['客户类型']=df['客户类型'].fillna(df['客户类型'].mode()[0])
    df['注册时间']=df['注册时间'].fillna(method='ffill')
    df['消费金额']=df['消费金额'].fillna(df['消费金额'].median())
    df['所在城市']=df['所在城市'].fillna('未知')
    df.to_csv(output_path, index=False)
    #TODO

if __name__ == '__main__':
    clean_csv_file('./customer_data.csv', './customer_data_cleaned.csv')
```
