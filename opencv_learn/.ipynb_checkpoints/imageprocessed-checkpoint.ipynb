```python
traslation_image = np.float32([[1,0,0],[0,1,0]]) #前x后y，前面1,0,0,1位置参数对不上是因为该处是单位矩阵，最后一列代表的是偏移量
traslated_image=cv2.warpAffine(image2,traslation_image,(w,h))
cv2.imshow('translation',traslated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

图像边缘检测

边缘检测是一种常用的图像处理技术，用于检测图像中的边缘，最常用的方法是 Canny 边缘检测。

Canny 边缘检测：edges = cv2.Canny(img, 100, 200),两个数值分别是x与y方向的梯度方向检测力度，像素之间要达到所设置的数值才会被检测到边缘

```python
edges = cv2.Canny(image, 1, 1)
cv2.imshow('edges',edges)
cv2.waitKey()
cv2.destroyAllWindows()
```

形态学操作：形态学操作常用于二值图像的处理，常见的操作有腐蚀、膨胀、开运算、闭运算等。

腐蚀（Erosion）：将图像中的白色区域收缩


kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5)) 数字依旧方向收缩程度


eroded_img = cv2.erode(img, kernel, iterations=1)

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,7))
eroded_img = cv2.erode(image, kernel, iterations=1)
cv2.imshow('eroded_img', eroded_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

膨胀（Dilation）：将图像中的白色区域扩展。

dilated_img = cv2.dilate(img, kernel, iterations=1)

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (8,5))
dilated_img = cv2.dilate(image, kernel, iterations=1)
cv2.imshow('dilated_img', dilated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

开运算与闭运算：

开运算（先腐蚀再膨胀）：用于去除小物体：pening_img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)


闭运算（先膨胀再腐蚀）：用于填补图像中的小孔洞：closing_img = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,23))
pening_img = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
cv2.imshow('pening_img', pening_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (27,13))
closing_img = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
cv2.imshow('closing_img', closing_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

图像轮廓检测:OpenCV 提供了强大的轮廓检测功能，可以用于对象识别、图像分割等应用。

检测轮廓：

```python
new_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
_, threshold_img = cv2.threshold(new_gray,110, 150, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(threshold_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
result = image.copy()   # 原彩色图复制一份
cv2.drawContours(result, contours, -1, (0, 255, 0), 2)
cv2.imshow('threshold_img', threshold_img)
cv2.imshow('contours', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

视频处理

```python
cap = cv2.VideoCapture('m2-res_360p.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # 处理每一帧
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
